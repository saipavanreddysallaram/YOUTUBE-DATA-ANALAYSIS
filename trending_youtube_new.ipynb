```python
import ipywidgets as widgets
```

```python
c =widgets.Dropdown(
    options=da['Gener'].unique(),
    value='Sports',
    description='Number:',
    disabled=False,
    )
def graph(a):
  cat=da[da['Gener']==a]
  newcat=cat.groupby(by='channelTitle').sum().sort_values(by='view_count',ascending=False).head(5)
  sns.barplot(x=newcat.index,y=newcat['view_count'])
  plt.xticks(rotation=45)
  plt.title(a)
  plt.show()
```

```python
widgets.interact(graph,a=c)
```
